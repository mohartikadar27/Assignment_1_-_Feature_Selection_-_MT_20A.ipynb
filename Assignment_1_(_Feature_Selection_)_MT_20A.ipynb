**1. Setting up the environment and importing libraries python Copy**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from keras.datasets import mnist
```

**2. Loading and Preprocessing MNIST Dataset**

```python
(X_train, y_train), (_, _) = mnist.load_data()
X = X_train.reshape(X_train.shape[0], -1).astype(np.float32) / 255.0  # Flatten and normalize
y = y_train
```

**3. t-SNE Visualization**

```python
# Apply t-SNE for dimensionality reduction and visualization
print("Applying t-SNE...")

# Ensure the data is in the right shape
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")

# Initialize t-SNE
tsne = TSNE(n_components=2, random_state=42, init='random', learning_rate='auto')

# Fit and transform a subset of data (10,000 samples to save time)
X_tsne = tsne.fit_transform(X[:10000])  # Use a subset to save time

# Plot the t-SNE result
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y[:10000], cmap='tab10', alpha=0.7)
plt.colorbar(scatter, ticks=range(10), label="Digit Class")
plt.title("t-SNE Visualization of MNIST")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```

**4. PCA Visualization**

```python
# Applying PCA for dimensionality reduction and visualization
print("Applying PCA...")

# Initialize PCA
pca = PCA(n_components=2)

# Fit and transform a subset of data (10,000 samples)
X_pca = pca.fit_transform(X[:10000])

# Plot the PCA result
plt.figure(figsize=(10, 8))
scatter_pca = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y[:10000], cmap='tab10', alpha=0.7)
plt.colorbar(scatter_pca, ticks=range(10), label="Digit Class")
plt.title("PCA Visualization of MNIST")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
